```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# DR4 exoplanet catalog

```python
# load catalog
dataframe = pd.read_csv('DR4_mock_exoplanet_catalog.csv', encoding='utf-8')
print(dataframe.shape)
```

```python
print(dataframe.shape)
```

```python
print(dataframe.keys())
```

```python
# get true, best-fit, median, 16th, and 84th percentile Porb values
period_true = np.array(dataframe['True period [days]'])
period_bestfit = np.array(dataframe['Best-fit period [days]'])
period_median = np.array(dataframe['MCMC period 50th [days]'])
period_16th = np.array(dataframe['MCMC period 16th [days]'])
period_84th = np.array(dataframe['MCMC period 84th [days]'])

print(np.sum(period_16th > period_84th)) # must be zero
print(np.sum(period_median > period_84th)) # must be zero
print(np.sum((period_16th < period_true) & (period_true < period_84th))/len(period_true)) # expect ~0.68
print(np.sum((period_16th < period_bestfit) & (period_bestfit < period_84th))/len(period_true)) # no firm expectation
```

Good! Nothing terrible has gone wrong with P16th and P84th, the uncertainties are realistic, and the best-fit values usually lie in the one-sigma range.

```python
# get true, best-fit, median, 16th, and 84th percentile mp values
mass_true = np.array(dataframe['True planet mass [M_J]'])
mass_bestfit = np.array(dataframe['Best-fit planet mass [M_J]'])
mass_median = np.array(dataframe['MCMC planet mass 50th [M_J]'])
mass_16th = np.array(dataframe['MCMC planet mass 16th [M_J]'])
mass_84th = np.array(dataframe['MCMC planet mass 84th [M_J]'])

print(np.sum(mass_16th > mass_84th)) # must be zero
print(np.sum(mass_median > mass_84th)) # must be zero
print(np.sum((mass_16th < mass_true) & (mass_true < mass_84th))/len(mass_true)) # expect ~0.68
print(np.sum((mass_16th < mass_bestfit) & (mass_bestfit < mass_84th))/len(mass_true)) # no firm expectation
```

For the planet masses, the uncertainties seem slightly underestimated. The best-fit values still typically fall in the one-sigma range.

```python
# get true, best-fit, median, 16th, and 84th percentile inclination values
inc_true = np.array(dataframe['True inclination [deg]'])
inc_bestfit = np.array(dataframe['Best-fit inclination [deg]'])
inc_median = np.array(dataframe['MCMC inclination 50th [deg]'])
inc_16th = np.array(dataframe['MCMC inclination 16th [deg]'])
inc_84th = np.array(dataframe['MCMC inclination 84th [deg]'])

print(np.sum(inc_16th > inc_84th)) # must be zero
print(np.sum(inc_median > inc_84th)) # must be zero
print(np.sum((inc_16th < inc_true) & (inc_true < inc_84th))/len(inc_true)) # expect ~0.68
print(np.sum((inc_16th < inc_bestfit) & (inc_bestfit < inc_84th))/len(inc_true)) # no firm expectation
```

```python
# get true, best-fit, median, 16th, and 84th percentile eccentricity values
ecc_true = np.array(dataframe['True eccentricity'])
ecc_bestfit = np.array(dataframe['Best-fit eccentricity'])
ecc_median = np.array(dataframe['MCMC eccentricity 50th'])
ecc_16th = np.array(dataframe['MCMC eccentricity 16th'])
ecc_84th = np.array(dataframe['MCMC eccentricity 84th'])

print(np.sum(ecc_16th > ecc_84th)) # must be zero
print(np.sum(ecc_median > ecc_84th)) # must be zero
print(np.sum((ecc_16th < ecc_true) & (ecc_true < ecc_84th))/len(ecc_true)) # expect ~0.68
print(np.sum((ecc_16th < ecc_bestfit) & (ecc_bestfit < ecc_84th))/len(ecc_true)) # no firm expectation
```

For the inclinations and eccentricities, the uncertainties are also slightly underestimated. This isn't too surprising, but it is good to keep in mind!

# Binary impostor catalog (DR5)

```python
# load catalog
dataframe = pd.read_csv('DR5_mock_planet_impostor_catalog.csv', encoding='utf-8')
print(dataframe.shape)
```

```python
print(dataframe.keys())
```

```python
# get primary, secondary, and apparent masses
primary_masses = np.array(dataframe['Primary stellar mass [M_\odot]'])
secondary_masses = np.array(dataframe['Secondary stellar mass [M_\odot]'])
apparent_masses = np.array(dataframe['Apparent stellar mass [M_\odot]'])
```

```python
# plot secondary mass versus primary mass
plt.figure(figsize=(8,5),dpi=250)
plt.scatter(primary_masses, secondary_masses, c='k', s=0.1)
plt.xlabel('Primary mass $M_1$ [$M_\odot$]')
plt.ylabel('Secondary mass $M_2$ [$M_\odot$]')
plt.show()
```

The planet impostor binaries have componenets with nearly exact masses, as expected.

```python
# plot distribution of 'apparent mass' to 'primary mass' ratios
plt.figure(figsize=(8,5),dpi=250)
plt.hist(apparent_masses/primary_masses, histtype='step', color='k', lw=2.5, bins=75, range=(0.9, 2.0), density=True)
plt.ylabel('Normalized $N$')
plt.xlabel('Ratio of apparent mass to primary mass: $M_\mathrm{apparent}/M_1$')
plt.show()
```

When a planet impostor binary is misinterpreted as a single star with a planet, the 'apparent' stellar mass is overestimated by ~10-50%.
